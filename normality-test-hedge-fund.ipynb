```python
def skewness(r):
    'measurement of the distortion of symmetrical distribution or asymmetry in a data set'
    'If skewness = 0, then Hedge fund is normaly distributed'
    demeaned_r = r - r.mean()
    # use the population standard deviation, so set dof=0
    sigma_r = r.std(ddof=0)
    exp = (demeaned_r**3).mean()
    return exp/sigma_r**3
```

```python
def kurtosis(r):
    'measure of the combined weight of a distributions tails relative to the center of the distribution curve'
    'if kurtosis = 3, then Hedge fund is normaly distributed'
    demeaned_r = r - r.mean()
    # use the population standard deviation, so set dof=0
    sigma_r = r.std(ddof=0)
    exp = (demeaned_r**4).mean()
    return exp/sigma_r**4
```
