```python
import pandas as pd
import datetime
from datetime import date, timedelta
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import matplotlib.pyplot as plt
pio.templates.default = "plotly_white"
```

```python
df = pd.read_csv("Tesla.csv")
print(df.head)
```

# What contains the data ?

1. Date
2. Opening price of the day
3. Higest price of the day
4. Lowest price of the day
5. Closing price of the day
6. Adjusted closing price of the day
7. Daily Volume

```python
plt.figure(figsize=(12, 6))
df['Open'].plot()
df['Close'].plot()
plt.ylabel(None)
plt.xlabel(None)
plt.title("Opening & Closing Price of Tesla")
plt.legend(['Open Price', 'Close Price'])
plt.show()
```

```python
figure = px.bar(df, 
                x = "Date", 
                y= "Close", 
                color="Close")
figure.update_xaxes(rangeslider_visible=True)
figure.show()
```

```python
figure = px.bar(df, x = "Date", y= "Close", color="Close")
figure.update_xaxes(rangeslider_visible=True)
figure.update_layout(title = "Tesla Stock Prices Over the Years", 
                     xaxis_rangeslider_visible=False)
figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=2, label="2y", step="year", stepmode="backward"),
            dict(step="all")])
    )
)
figure.show()
```

Price changes throughout the years:


```python
df["Date"] = pd.to_datetime(df["Date"])
df['Year'] = df['Date'].dt.year
df["Month"] = df["Date"].dt.month
fig = px.line(df, 
              x="Month", 
              y="Close", 
              color='Year', 
              title="Price changes in Tesla")
fig.show()
```
